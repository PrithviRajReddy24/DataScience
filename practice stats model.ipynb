```python
from sklearn.datasets import load_boston
```

```python
b_data = load_boston()
```

```python
b_data.keys()
```

```python
print(b_data.DESCR)
```

```python
import pandas as pd
boston_data = pd.DataFrame(b_data.data,columns=b_data.feature_names)
boston_data['MEDV'] = b_data.target
```

```python
boston_data.describe()
```

```python
boston_data.shape
```

```python
boston_data.info()
```

```python
boston_data.head()
```

```python
boston_data.MEDV.corr(boston_data.AGE)
```

```python
boston_data.corr()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(boston_data.corr(),annot=True,cmap='Accent')
```

```python
from statsmodels.formula.api import ols
mod = ols("MEDV~RM",data=boston_data).fit()
```

```python
mod.summary()
```

```python
ft = ' + '.join(boston_data.drop(['MEDV','CHAS'],axis=1).columns)
mod = ols('MEDV~'+ft,data=boston_data).fit()
```
